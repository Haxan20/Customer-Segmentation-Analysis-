```python
#Let prepare and Preprocess our data
customers[['Income', 'SpendingScore']].describe().round(2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
customers_scaled = scaler.fit_transform(customers[['Income', 'SpendingScore']])
```

```python
customers_scaled = pd.DataFrame(customers_scaled, columns = ['Income', 'SpendingScore'])
```

```python
customers_scaled.describe().round(2)
```

```python
#let create our kmeans model
from sklearn.cluster import KMeans
km = KMeans(n_clusters = 3, n_init = 25, random_state = 1234)
```

```python
km.fit(customers_scaled)
```

```python
km.labels_
```

```python
km.inertia_
```

```python
#Let Evaluate the model and visualize it
pd.Series(km.labels_).value_counts().sort_index()
```

```python
km.cluster_centers_
```

```python
cluster_centers = pd.DataFrame(km.cluster_centers_, 
                               columns = ['Income', 'SpendingScore'])
cluster_centers
```

```python
#Let Set the figure size
plt.figure(figsize=(8, 6))

#Plot the Clusters
ax = sns.scatterplot(data = customers_scaled,
                     x = 'Income',
                     y = 'SpendingScore', 
                     hue = km.labels_, 
                     palette = 'colorblind', 
                     alpha = 0.8, 
                     s = 150,
                     legend = False)

#let plot the Centroids and add the centroid labels for storytelling purpose
ax = sns.scatterplot(data = cluster_centers, 
                     x = 'Income',
                     y = 'SpendingScore', 
                     hue = cluster_centers.index, 
                     palette = 'colorblind', 
                     s = 600,
                     marker = 'D',
                     ec = 'black', 
                     legend = False)

#Let include the Centroid Labels
for i in range(len(cluster_centers)):
               plt.text(x = cluster_centers.Income[i], 
                        y = cluster_centers.SpendingScore[i],
                        s = i, 
                        horizontalalignment='center',
                        verticalalignment='center',
                        size = 15,
                        weight = 'bold',
                        color = 'white')
```

```python
#it obvious that members of our second cluster does not have thesame characteristics
#Let analyzed the right number of clusters to choose using the within cluster sum of squares(wcss)
wcss = []
for k in range(2, 11):
    km = KMeans(n_clusters = k, n_init = 25, random_state = 1234)
    km.fit(customers_scaled)
    wcss.append(km.inertia_)

wcss_series = pd.Series(wcss, index = range(2, 11))

plt.figure(figsize=(8, 6))
ax = sns.lineplot(y = wcss_series, x = wcss_series.index)
ax = sns.scatterplot(y = wcss_series, x = wcss_series.index, s = 150)
ax = ax.set(xlabel = 'Number of Clusters (k)', 
            ylabel = 'Within Cluster Sum of Squares (WCSS)')
```

```python
km = KMeans(n_clusters = 5, n_init = 25, random_state = 1234)
km.fit(customers_scaled)
```

```python
cluster_centers = pd.DataFrame(km.cluster_centers_, columns = ['Income', 'SpendingScore'])
```

```python
#Let set the figure size
plt.figure(figsize=(8, 6))

#Plot the Clusters
ax = sns.scatterplot(data = customers_scaled,
                     x = 'Income',
                     y = 'SpendingScore', 
                     hue = km.labels_, 
                     palette = 'colorblind', 
                     alpha = 0.8, 
                     s = 150,
                     legend = False)

#let plot the Centroids and add centroid labels
ax = sns.scatterplot(data = cluster_centers, 
                     x = 'Income',
                     y = 'SpendingScore', 
                     hue = cluster_centers.index, 
                     palette = 'colorblind', 
                     s = 600,
                     marker = 'D',
                     ec = 'black', 
                     legend = False)


# Add Centroid Labels
for i in range(len(cluster_centers)):
               plt.text(x = cluster_centers.Income[i], 
                        y = cluster_centers.SpendingScore[i],
                        s = i, 
                        horizontalalignment='center',
                        verticalalignment='center',
                        size = 15,
                        weight = 'bold',
                        color = 'white')
```

```python
customers['Cluster'] = km.labels_.tolist()
customers.head(10)
```

```python
#Let dummy code our gender to convert the categorical data into numerical data
customers = pd.get_dummies(customers, columns = ['Gender'])
customers.head(10)
```

```python
customers.agg({
    'Gender_Female': 'mean',
    'Gender_Male': 'mean',
    'Age': 'median',
    'Income': 'median',
    'SpendingScore': 'median'}).round(2)
```

```python
customers.groupby('Cluster').agg({
    'Gender_Female': 'mean',
    'Gender_Male': 'mean',
    'Age': 'median',
    'Income': 'median',
    'SpendingScore': 'median'}).round(2)
```
